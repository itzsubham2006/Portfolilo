```python
#breast cancer model

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pickle
import sklearn
```

```python
data = pd.read_csv("C:/Users/Subham Pathak/Desktop/AI/PORT_FOLIO/data_sets/breast_cancer_data.csv")
print(data)
```

```python
cols_to_drop = [
    'fractal_dimension_mean', 'fractal_dimension_se', 'symmetry_se',
    'smoothness_se', 'smoothness_mean', 'fractal_dimension_worst',
    'symmetry_mean', 'concave points_se', 'compactness_se', 'concavity_se',
    'radius_se', 'perimeter_se', 'area_se', 'symmetry_worst',
    'texture_mean', 'compactness_mean', 'smoothness_worst', 'perimeter_mean'
]

data = data.drop(columns = cols_to_drop, errors='ignore')
data = data.iloc[:, :-1]
data.head(30)
```

```python
x = data.drop(['id', 'diagnosis'], axis=1)
y = data['diagnosis']
print(x)
```

```python
# Data standardization

scalar = StandardScaler()
```

```python
# splitting the data into train and test datas

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42, stratify = y)
model2 = RandomForestClassifier(max_depth=5)
```

```python
model2.fit(x_train, y_train)
pickle.dump(model2, open('model2.pkl', 'wb'))
predicted = model2.predict(x_test)
acc = accuracy_score(y_test, predicted)
```
