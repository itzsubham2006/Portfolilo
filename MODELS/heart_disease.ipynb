```python
#importing the dependecies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import sklearn
import pickle
```

```python
data1 = pd.read_csv(r"C:\Users\Subham Pathak\Desktop\AI\PORT_FOLIO\data_sets\heart.csv")
x = data1.iloc[:,:-1]
y = data1['target']
```

```python
# splitting into train test split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 42)
```

```python
model = RandomForestClassifier()
model.fit(x_train, y_train)
pickle.dump(model, open('model.pkl', 'wb'))
```

```python

def user():
    
    user_input = (46,0,2,142,177,0,0,160,1,1.4,0,0,2)
    input_data_arr = np.asarray(user_input)
    reshaped = input_data_arr.reshape(1,-1)
    
    prediction = model.predict(reshaped)
    if(prediction == 1):
        print("The person is in danger ")
    else:
        print("The person is safe.")
        
user()
```
